### Fourier Series

```python
# CAS questins from advanced mathematics for engineers
#computing partial sums
def partial_sum(a_n,b_n,w,n,t,sums = 0):
    n = var('n')
    t = var('t')
    func(n,t) = a_n*cos(w*t) + b_n*sin(w*t)
    return_list =[]
    for i in range(1,sums+1):
        sum_ = 0
        for j in range(1,i+1):
            sum_ += func(j,t)
        return_list.append(sum_)
    return return_list
```

```python
# example 1 pg 484 ADVANCED ENGINEERING MATHEMATICS BY ERWIN ERESZIG
var('n t')
a(n) = (2/(n*pi))*sin(n*pi/2)
b(n) = 0
w(n) = n*pi/2
l = partial_sum(a,b,w,n,t,sums = 10)
plot(l, legend_label = ['S'+str(i) for i in range(1,len(l)+1)] , linestyle = .dashed')
```
