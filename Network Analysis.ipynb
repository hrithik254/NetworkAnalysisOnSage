```python
# CAS questions from advanced mathematics for engineers
#computing partial sums
def partial_sum(a_n,b_n,w,n,t,sums = 0):
    n = var('n')
    t = var('t')
    func(n,t) = a_n*cos(w*t) + b_n*sin(w*t)
    return_list =[]
    for i in range(1,sums+1):
        sum_ = sum(func(n,t),n,1,i)
        return_list.append(sum_)
    return return_list

def animate_fourier(list_of_partial_sums):
    animation = animate(list_of_partial_sums)
    animation.show()
```

```python
def laplace(function, domain_variable, simplify = True ):
    """domain variable is the variable with respect to integration would take place
    simplify argument converts the transformed function into partial fractions. The function thus returned has 
    arguments in alphabetical order ."""
    var('s')
    assume(s>0)
    transformed_function = integral(function*exp(-s*domain_variable),domain_variable,0,oo)
    if simplify:
        return transformed_function.partial_fraction()
    else:
        return transformed_function
```

```python
def z_to_y(z):
    d = z.det()
    y = matrix(QQ,2) #important to write QQ otherwise it assumes integers
    y[0,0] =  z[1,1]/d
    y[0,1] = -z[0,1]/d
    y[1,0] = -z[1,0]/d
    y[1,1] =  z[0,0]/d
    return y

def z_to_h(z):
    d = z.det()
    h = matrix(QQ,2)
    h[0,0] = d/z[1,1]
    h[0,1] = z[0,1]/z[1,1]
    h[1,0] = -z[1,0]/z[1,1]
    h[1,1] = 1/z[1,1]
    return h

def z_to_t(z):
    d = z.det()
    t = matrix(QQ,2)
    t[0,0] = z[0,0]/z[1,0]
    t[0,1] = d/z[1,0]
    t[1,0] = 1/z[1,0]
    t[1,1] = z[1,1]/z[1,0]
    return t

a = matrix([[5/6,1/2],[1/2,5/2]])
a = matrix(RR,a)

```

```python
def z_to_z(z):
    z = matrix(SR,z)
    return z

def z_to_y(z):
    d = z.det() 
    y = matrix(SR,2)                        #important to write SR (symbolic r) otherwise it assumes integers
    y[0,0] =  z[1,1]/d
    y[0,1] = -z[0,1]/d
    y[1,0] = -z[1,0]/d
    y[1,1] =  z[0,0]/d
    return matrix(y)

def z_to_h(z):
    d = z.det()
    h = matrix(SR,2)
    h[0,0] = d/z[1,1]
    h[0,1] = z[0,1]/z[1,1]
    h[1,0] = -z[1,0]/z[1,1]
    h[1,1] = 1/z[1,1]
    return h

def z_to_t(z):
    d = z.det()
    t = matrix(SR,2)
    t[0,0] = z[0,0]/z[1,0]
    t[0,1] = d/z[1,0]
    t[1,0] = 1/z[1,0]
    t[1,1] = z[1,1]/z[1,0]
    return t
        
class two_port():
    
    def __init__(self,z,mode = 'z'):
        if mode == 'z':
            self.z = z_to_z(z)
            self.y = z_to_y(z)
            self.h = z_to_h(z)
            self.t = z_to_t(z)
            
class ideal_transformer(two_port):
    def __init__(self,n):
        self.t = matrix([[n,0],[0,1/n]])
        
class gyrator(two_port):
    def __init__(self,n):
        self.t = matrix([[0,n],[1/n,0]])
        
class NIC(two_port):
    def __init__(self,n):
        self.h = matrix([[0,n][1/n,0]])
```
