```python
t = var("t")
x = function("x")(t)
```

```python
# example 4.1
f = desolve(diff(x,t,2) + 5*diff(x,t) + 4*x, dvar = x, ics =[0,2,-1], show_method = True)
show("x(t) = ", f[0], ", method used =  ", f[1])
plot_ = plot(f[0],x,-1,4,axes_labels=['$t$ axis','$x(t)$ axis'])
plot_ +=  text("${}$".format(latex(f[0])),(2,-4), fontsize = 25,bounding_box={'boxstyle':'round', 'fc':'w'})
plot_
```

```python
#example 4.5 
f = desolve(diff(x,t,2) + 3*diff(x,t) + 2*x - 4*exp(t), dvar = x, ics =[0,1,-1], show_method = True)
show("x(t) = ", f[0], ", method used =  ", f[1])

plot_ = plot(f[0],axes_labels=['$t$ axis','$x(t)$ axis'])
plot_ +=  text("${}$".format(latex(f[0])),(0.8,6), fontsize = 25,bounding_box={'boxstyle':'round', 'fc':'w'})
plot_
```
